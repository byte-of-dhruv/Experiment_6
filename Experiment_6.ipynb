```python
#Question 6 Write a python program to calcultare the gross salary of an employee based
#if basic<=10000 : HRA = 20% , DA = 80%
#if basic<=20000 : HRA = 25% , DA = 90%
#otherwise : HRA = 30% , DA = 95%

basic_salary = float(input("Enter the basic salary: "))

if basic_salary <= 10000:
    hrA = 0.20 * basic_salary
    Da = 0.80 * basic_salary
elif basic_salary <= 20000:
    hrA = 0.25 * basic_salary
    Da = 0.90 * basic_salary
else:
    hrA = 0.30 * basic_salary
    Da = 0.95 * basic_salary

gross_salary = basic_salary + hrA + Da

print(f"Basic Salary: {basic_salary}")
print(f"HRA: {hrA}")
print(f"DA: {Da}")
print(f"Gross Salary: {gross_salary}")
```

```python
# Write a python program to calculate income tax based on annual income
#upto $2,50,000:No tax
#$2,50,001-$5,00,000:5%
#$5,00,001-$10,00,000:20%
#above $10,00,000:30%
annual_income = float(input("Enter the annual income: "))
if annual_income <=250000:
    tax = 0
    print("No tax")
elif annual_income <=500000:
    tax = 0.05 * (annual_income -250000)
    print("Tax: ",tax)
elif annual_income<=1000000:
    tax = (250000 * 0.05)+0.2*(annual_income-500000)
    print("Tax: ", tax)
else:
  tax = (250000 * 0.05)+0.2*(500000)+0.3*(annual_income-1000000)
  print("Tax: ", tax )
```

```python
date = input("Enter the date in format dd/mm/yyyy: ")
dd, mm, yy = map(int, date.split('/'))

# Check valid month
if mm < 1 or mm > 12:
    print("Date is invalid.")

else:
    # Days in month
    if mm in [1, 3, 5, 7, 8, 10, 12]:
        max_days = 31
    elif mm in [4, 6, 9, 11]:
        max_days = 30
    else:
        # February
        if (yy % 4 == 0 and yy % 100 != 0) or (yy % 400 == 0):
            max_days = 29
        else:
            max_days = 28

    # Check valid day
    if dd < 1 or dd > max_days:
        print("Date is invalid.")

    else:
        # Increment date
        if dd == max_days:
            dd = 1
            if mm == 12:
                mm = 1
                yy += 1
            else:
                mm += 1
        else:
            dd += 1

        print("The incremented date is:", dd, mm, yy)


```
